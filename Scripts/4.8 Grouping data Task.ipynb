```python
# Question 4:  Create a loyalty flag for existing customers using the transform() and loc() functions

# First aggregating using the transform function and creating a new column 'max_order' 

df_ords_prods_merge['max_order'] = df_ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# Secondly, using the loc() function to create new column 'loyalty_flag' based on some conditions

df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
df_ords_prods_merge.loc[(df_ords_prods_merge['max_order'] <= 40) & (df_ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Checking output
df_ords_prods_merge.head()
```

```python
# Checking frequency of the loyalty_flag column
df_ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
# Question 5: Use the loyalty flag you created and check the basic statistics of the product prices for each loyalty category

df_ords_prods_merge.groupby('loyalty_flag').agg({'prices' : ['mean','min', 'max']})
```

```python
# Question 6: Create a spending flag for each user based on the average price across all their orders using the following criteria:
# If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”
# If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”

# firstly, use the transform() function to create a new column 'average_price' 
df_ords_prods_merge['average_price'] = df_ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Checking output
df_ords_prods_merge.head()
```

```python
# Secondly, use the .loc() function to create a new column 'spending_flag' based on given conditions
 
df_ords_prods_merge.loc[df_ords_prods_merge['average_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['average_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
# Checking output
df_ords_prods_merge.head()
```

```python
# checking frequency of spending_flag column

df_ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

```python
# checking list of columns
df_ords_prods_merge.columns
```

```python
# Question 7
# Create an order frequency flag that marks the regularity of a user’s ordering behavior according 
# to the median in the “days_since_prior_order” column. The criteria for the flag should be as follows:

# If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”
# If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”
# If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”
```

```python
# firstly, use the transform() function to create a new column 'median_days_since_prior_order'

df_ords_prods_merge['median_days_since_prior_order'] = df_ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# Checking output
df_ords_prods_merge.head()
```

```python
# Checking shape

df_ords_prods_merge.shape
```

```python
# Secondly, use the .loc() function to create a new column 'frequency_flag' based on given conditions


df_ords_prods_merge.loc[df_ords_prods_merge['median_days_since_prior_order'] > 20, 'frequency_flag'] = 'Non-frequent customer'
```

```python
df_ords_prods_merge.loc[(df_ords_prods_merge['median_days_since_prior_order'] > 10) & (df_ords_prods_merge['median_days_since_prior_order'] <= 20), 'frequency_flag'] = 'Regular customer'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['median_days_since_prior_order'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
# Checking output
df_ords_prods_merge.head()
```

```python
# Checking shape
df_ords_prods_merge.shape
```

```python
# Checking value count of the frequency_flag column

df_ords_prods_merge['frequency_flag'].value_counts(dropna = False)
```

#### If-statement with user-defined function

```python
# Defining function

def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# Applying function (axis = 0 is for rows, conversely axis = 1 is for columns)

df['price_range'] = df.apply(price_label, axis = 1)
```

```python
df['price_range'].value_counts(dropna = False)
```

```python
# Checking for most expensive product in the subset
df['prices'].max()
```

#### if-statement with the .loc function

```python
# Using the .loc function

df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
df['price_range_loc'].value_counts(dropna = False)
```

#### I will categorize the values in the 'order_hour_of_day' column based on their frequency;

##### <10,000 (fewest orders)
##### >10,000 to 2,000,000 (regular orders)
##### >2,000,000 (most orders)

##### see loop below

```python
# Creating loop

results1 = []

for value in df_ords_prods_merged['order_hour_of_day']:
    if value >=9 and value <=17:
        results1.append('Most orders')
    elif value >=2 and value <=5:
        results1.append('Fewest orders')
    else:
        results1.append('Regular orders')
```

```python
results1
```

```python
# creating new column 'busiest_period_of_day'.

df_ords_prods_merged['busiest_period_of_day'] = results1
```
